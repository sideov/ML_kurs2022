```python
data["end_station_id"] = data["all_stations_id"].apply(lambda x: x[1:-1].split(",")[-1])
data["end_station_nm"] = data["all_stations_nm"].apply(lambda x: list(map(lambda x: x[1:-1], x.strip("[]").split(",")))[-1])
```

```python
time_s = pd.Series(dtype = "float")
```

```python
for i, station_from in enumerate(list(data["start_station_nm"].unique())):
    for station_to in list(data["start_station_nm"].unique())[i:]:
        
        stations_time = data.loc[((data["end_station_nm"] == station_to)  & 
                                      (data["start_station_nm"] == station_from)), ("route_dur")]
        
        station_opt_time = min(list(stations_time.values))
        
        d = pd.Series({f"from {station_from} to {station_to} opt time: ": station_opt_time})
        
        time_s = pd.concat([time_s, d], axis = 0)
        
        #time_s[f"from {station_from} to {station_to} opt time: "] = station_opt_time
        
    print(i)
        
        #print(station_from, station_to, station_opt_time)
```
