```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from datetime import datetime
import sklearn
```

```python
data = pd.read_csv(r'/Users/maximvasilev/Desktop/ML/about_stations.csv')
data
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data_new = data[["Arts & Entertainment", "College & University", "Event",
                "Food", "Nightlife Spot", "Outdoors & Recreation", "Professional & Other Places", "Residence",
               "Shop & Service", "Travel & Transport"]]
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaler
```

```python
min_max_scaler = min_max_scaler.fit(data_new[["Arts & Entertainment", "College & University", "Event",
                "Food", "Nightlife Spot", "Outdoors & Recreation", "Professional & Other Places", "Residence",
               "Shop & Service", "Travel & Transport"]])

min_max_scaler = min_max_scaler.transform(data_new[["Arts & Entertainment", "College & University", "Event",
                "Food", "Nightlife Spot", "Outdoors & Recreation", "Professional & Other Places", "Residence",
               "Shop & Service", "Travel & Transport"]])
min_max_scaler
```

```python
data_mod = data.copy()
```

```python
data_mod[["Arts & Entertainment", "College & University", "Event",
                "Food", "Nightlife Spot", "Outdoors & Recreation", "Professional & Other Places", "Residence",
               "Shop & Service", "Travel & Transport"]] = min_max_scaler
```
